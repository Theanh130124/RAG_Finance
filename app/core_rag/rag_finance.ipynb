#Extract PDF


```python
from langchain_community.document_loaders import DirectoryLoader, UnstructuredPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter


def load_pdf_files(data):
    loader = DirectoryLoader(
        path=data,
        glob="*.pdf",
        loader_cls=UnstructuredPDFLoader
    )
    documents = loader.load()
    return documents
```

```python
import json
from langchain.docstore.document import Document

def json_to_documents(json_data):
    docs = []
    for item in json_data:
        text = ""
        for key, value in item.items():
            text += f"{key}: {value}\n"
        docs.append(Document(page_content=text))
    return docs
#-> trả ra text trong docs
```

#Preprocessing (Tiền xử lý dữ liệu)

```python
import re

def preprocess_data(text):


    #Xóa URL
    text = re.sub(r'(https?://\S+|www\.\S+)', '', text)
    # Xoá các dòng chỉ chứa dấu = hoặc -
    text = re.sub(r'^[=\-]{2,}\s*$', '', text, flags=re.MULTILINE)
    #Xoá các ký tự bảng markdown (|, ---)
    text = re.sub(r'\|.*?\|', '', text)
    # Xoá emoji và ký tự Unicode không cần thiết
    text = re.sub(r'[^\w\s,.!?à-ỹÀ-Ỹ\-–]', '', text)
    # Xoá khoảng trắng thừa và dòng trống
    text = re.sub(r'\s+', ' ', text).strip()
    return text
```

```python
from langchain_core.documents import Document

cleaned_data = []
for doc in extracted_data:
    cleaned_content = preprocess_data(doc.page_content)
    cleaned_doc = Document(
        page_content=cleaned_content,
        metadata=doc.metadata
    )
    cleaned_data.append(cleaned_doc)
```

#Chunking với text

```python
def text_split(cleaned_data):
    text_splitter = RecursiveCharacterTextSplitter(
        chunk_size=512,
        chunk_overlap=20,
        separators=["\n\n", "\n", ". ", " ", ""]
    )
    text_chunks = text_splitter.split_documents(cleaned_data)
    return text_chunks
```

```python
text_chunks = text_split(cleaned_data)
print(f'So luong chunk da tach', len(text_chunks))
```

#Cross-encoder (Re-rank)

```python

#Tải model
from sentence_transformers import CrossEncoder

cross_encoder = CrossEncoder('itdainb/PhoRanker')

def rerank_documents(query, docs, top_n):

    # Chuyển đổi tài liệu thành định dạng phù hợp với cross-encoder
    pairs = [[query, doc.page_content] for doc in docs]

    # Tính điểm cho từng cặp (truy vấn, tài liệu)
    scores = cross_encoder.predict(pairs)

    # Kết hợp điểm với tài liệu ,Sắp xếp tài liệu theo điểm số giảm dần
    reranked = sorted(zip(docs, scores), key=lambda x: x[1], reverse=True)

    return [doc for doc, _ in reranked[:top_n]]
```

#MEMORY

```python
from langchain.memory import ConversationBufferMemory
from langchain.chains import ConversationalRetrievalChain
from langchain_core.prompts import MessagesPlaceholder
from langchain.schema import BaseMessage

# Khởi tạo memory
memory = ConversationBufferMemory(
    memory_key="chat_history",
    return_messages=True,
    output_key='answer'
)

```

```python
#prompt có thêm <system>

from langchain.chains import create_retrieval_chain
from langchain.chains.combine_documents import create_stuff_documents_chain
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder

system_prompt = (
    "Bạn là một chuyên gia tư vấn tài chính chuyên nghiệp, "
    "có nhiệm vụ hỗ trợ và tư vấn cho người dùng về các vấn đề tài chính dựa trên **thông tin ngữ cảnh được cung cấp** và **lịch sử hội thoại trước đó**. "
    "Chỉ sử dụng thông tin trong ngữ cảnh - không được suy đoán hoặc bịa thêm thông tin. "
    "Nếu ngữ cảnh không chứa thông tin cần thiết, hãy trả lời: 'Xin lỗi, tôi không có đủ thông tin chuyên môn để tư vấn về vấn đề này.'\n\n"

    "HƯỚNG DẪN TRẢ LỜI:\n"
    "1. Luôn trả lời bằng **tiếng Việt tự nhiên, chuyên nghiệp**\n"
    "2. Tập trung vào các lĩnh vực tài chính như: đầu tư, tiết kiệm, quản lý tài chính cá nhân, bảo hiểm, thuế, hưu trí\n"
    "3. Trình bày rõ ràng, ngắn gọn, tối đa 8-9 câu\n"
    "4. Nếu có thể, đưa ra lời khuyên thực tế và an toàn\n"
    "5. Luôn nhắc nhở về rủi ro khi đề cập đến đầu tư\n"
    "6. Không đưa ra lời khuyên đầu tư cụ thể nếu không có đủ thông tin\n\n"

    "=== LỊCH SỬ HỘI THOẠI ===\n{chat_history}\n\n"
    "=== THÔNG TIN NGỮ CẢNH TÀI CHÍNH ===\n{context}\n\n"
    "=== CÂU HỎI TƯ VẤN ===\n{input}"
)

prompt_with_memory = ChatPromptTemplate.from_messages([
    ("system", system_prompt),
    MessagesPlaceholder(variable_name="chat_history"),
    ("human", "{input}")
])
```

```python
# Tạo chain với memory
question_answer_chain_with_memory = create_stuff_documents_chain(llm, prompt_with_memory)
rag_chain_with_memory = create_retrieval_chain(retriever, question_answer_chain_with_memory)
```

```python
def chat_with_memory(query):
    inputs = {
        "input": query,
        "chat_history": memory.chat_memory.messages
    }

    try:
        response = rag_chain_with_memory.invoke(inputs)
    except Exception as e:
        print(f"Lỗi khi gọi rag_chain_with_memory: {e}")
        # fallback: không dùng memory
        try:
            response = rag_chain.invoke({"question": query})
        except Exception as e2:
            print(f"Lỗi fallback không dùng memory: {e2}")
            return {"answer": "Xin lỗi, hiện tại tôi không trả lời được."}

    # Lấy answer một cách an toàn
    answer = None
    if isinstance(response, dict):
        answer = response.get('answer') or response.get('result') or str(response)
    elif hasattr(response, 'content'):
        answer = response.content
    elif isinstance(response, str):
        answer = response
    else:
        answer = str(response)

    # Lưu vào memory (nếu có) - SỬA Ở ĐÂY: đổi "output" thành "answer"
    if answer:
        memory.save_context({"input": query}, {"answer": answer})  # Sửa "output" -> "answer"

    return {"answer": answer}


# Hàm tiện ích cho memory
def print_chat_history():
    """In lịch sử hội thoại"""
    print("\n" + "="*50)
    print("LỊCH SỬ HỘI THOẠI")
    print("="*50)
    for i, message in enumerate(memory.chat_memory.messages):
        role = "Human" if i % 2 == 0 else "AI"
        print(f"{role}: {message.content}")
    print("="*50)

def clear_memory():
    """Xóa memory"""
    memory.clear()
    print("Đã xóa lịch sử hội thoại")
```
